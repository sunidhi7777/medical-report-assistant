```python
qadata = {}
count = 1

while True:
    question = input("Enter your query or 'no' to end: ")
    if question.strip().lower() == "no":
        print(" \n thank you ")
        break

    if question.strip().lower().startswith("refer to q"):
        qid = question.strip().split()[-1].upper()
        if qid in qadata:
            print(f"\n {qid}:")
            print("Question:", qadata[qid]["question"])
            print("Answer:", qadata[qid]["answer"])
        else:
            print("ID not found.")
        continue

    response = client.models.generate_content(
        model="gemini-2.0-flash",
        contents=[
            """You are a knowledgeable and helpful doctor.
            You can analyze medical and neuroscience documents (PDFs) and provide clear answers based on them , keep it short in 2 line.
            If asked, you can suggest possible prescriptions, medicines, dosages, or treatment plans (only for educational and informative purposes).""",
            question,
            myfile
        ]
    )

    answer = response.text.strip()
    print("\n answer :")
    print(answer)

    qid = f"Q{count}"
    qadata[qid] = {"question": question, "answer": answer}
    print(f"\n saved as {qid}")
    count += 1
```
